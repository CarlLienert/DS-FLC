This command requires a bit more thought:

* The part inside brackets: `df_can_langs['lang_known']>=9025` is called a *boolean*.  That is, it is either True or False.  If it is True, the corresponding row is part of the output. If it is False, the corresponding row is **not** part of the output.
* `df_can_langs['lang_known']` specifies the column of interest.
* `>=9025` gives the criteria on the values of interest in that column
